# 第六课 数据分析工具Pandas高阶
## 第七节 数据规整 (1) -- 数据合并 concat

```python
import pandas as pd
import numpy as np
```

```python
# 创建dataframe
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])

df3 = pd.DataFrame({'C': ['A8', 'A9', 'A10', 'A11'],
                    'D': ['B8', 'B9', 'B10', 'B11'],
                    'E': ['C8', 'C9', 'C10', 'C11'],
                    'F': ['D8', 'D9', 'D10', 'D11']},
                    index=[0, 1, 2, 3])
```

```python
df1
```

```python
df2
```

```python
df3
```

* 纵向合并

```python
# 列名相同
pd.concat([df1, df2], axis=0)
```

```python
# 列名不同
pd.concat([df1, df3], axis=0)
```

* 横向合并

```python
# index相同
pd.concat([df1, df3], axis=1)
```

```python
# index不同
pd.concat([df1, df2], axis=1)
```
