# 第六课 数据分析工具Pandas高阶
## 第九节 数据规整(3) -- 数据重构stack、unstack

```python
import pandas as pd
```

```python
# 创建dataframe
header = pd.MultiIndex.from_product([['Semester1','Semester2'],['Maths','Science']])
d = [[12,45,67,56],[78,89,45,67],[45,67,89,90],[67,44,56,55]]
 
df = pd.DataFrame(d, index=['Alisa','Bobby','Cathrine','Jack'], columns=header)
df
```

* stack()

```python
stacked_df = df.stack()
stacked_df
```

```python
# level 参数
df.stack(level=0)
```

```python
df.stack(level=1)
```

```python
df.stack(level=-1)
```

* unstack()

```python
stacked_df
```

```python
stacked_df.unstack()
```

```python
stacked_df.unstack(level=0)
```

```python
stacked_df.unstack(level=1)
```

```python
stacked_df.unstack(level=-1)
```
