## 2-1 带参数的URL访问（爬取评论、爬取歌词）总结

### 总结
    思路：1：确定你所需要获取的数据是html的还是XHR(返回json数据)
        2：如果是html的就用美丽汤解决，如果是XHR，一定要找到其通用URL(即访问哪里、去哪)
        3：确定去哪后，要知道自己需要干什么，所以需要 Headers /Query string parameters 里的参数（当然你也可以直接定位，但是可读性太差！）
        4：再加一个Request Headers 来给自己的爬虫披上一件浏览器的马甲，伪装成浏览器可以有效规避反爬虫技术
        5：获取数据后清洗数据，正则化我用的还是不大好
    爬取评论：  本次实验，针对评论有两个参数在变化，一个是页码，还有一个是用户最后一个id
            其中用户前一页最后一个id应该是起到，读取一个页面内所有用户的作用(从上到下按顺序排)
    爬取歌词：  和评论最大的不同点在于，虽然也是在XHR里，但是每个歌词变化的只有music_id，所以只要不断更新music_id就能爬取不同的歌词
            但是，如何寻找musicid呢？我们在上一页的XHR文件里找到了每首歌的id
            说白了，最终目的还是为了定位所需要资源的url
