```python
# 本地Chrome浏览器设置方法
from selenium import webdriver # 从selenium库中调用webdriver模块
import time # 调用time模块
import cv2
import requests
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
driver = webdriver.Chrome()
url_login = 'https://toefl.neea.edu.cn'
password = 'Yuyue123'
student_name = '21005090' 
driver.get(url_login) # 访问页面
time.sleep(5) # 暂停两秒，等待浏览器缓冲
login_2 = driver.find_element_by_class_name('footer_butt')
login_2.click()

def login(student_name,password):
    student_login = driver.find_element_by_id('userName')
    student_login.send_keys(student_name)
    user_password = driver.find_element_by_id('password')
    user_password.send_keys(password)
    validation = driver.find_element_by_id('verifyCode')
    validation.send_keys(input())
    login = driver.find_element_by_id('btnLogin')
    login.click()
```
