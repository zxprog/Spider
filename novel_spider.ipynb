```python
import requests 
import re
url = 'https://read.dmnico.cn/book/401348.html'
response = requests.get(url)#利用get策略发送请求
response.encoding = 'utf-8'#设置编码方式，即用中文展现
html = response.text #返会改网页的html架构信息
print(html)


```

```python
#正则化中 .表示匹配任意可见字符 *表示匹配任意数量？表示匹配最短的字符个数 这里是在找章节的超链接
ul = re.findall(r'<ul class="list-unstyled row text-center ff-playurl-line ff-playurl" data-active="401348--" data-more="34">.*</a></li></ul>',html,re.S)[0]
#print(ul)
#title = re.findall(r'<ul class="list-unstyled row text-center ff-playurl-line ff-playurl" data-active="401348--" data-more="34">.*? </ul>',html)[0]
#print(title)
chapter_info_list=re.findall(r'href="(.*?)"(.*?)>(.*?)<',ul,re.S)

for chapter_info in chapter_info_list[:10]:  #测试
    chapter_url,_,chapter_title = chapter_info
    #chapter_url = "https://read.dmnico.cn/book/401348.html%s"%chapter_url#用+号相连的话，会不停生成新的字符串变量，占内存
    print(chapter_url,chapter_title)
    chapter_response = requests.get(chapter_url)
#小说标题    
title = re.findall(r'<meta name="keywords" content="(.*?),(.*?)">',html)[0][0]
print(title)
```

```python

for chapter_info in chapter_info_list[:1]:
    chapter_url,_,chapter_title = chapter_info
    #chapter_url = "https://read.dmnico.cn/book/401348.html%s"%chapter_url#用+号相连的话，会不停生成新的字符串变量，占内存
    chapter_response = requests.get(chapter_url)
    chapter_response.encoding = 'utf-8'
    chapter_html = chapter_response.text
    #提取章节内容
    chapter_content = re.findall(r'<div class="read-content j_readContent">(.*?)</div>',chapter_html,re.S)[0]
    #清洗数据
    chapter_content = chapter_content.replace(' ','')
    chapter_content = chapter_content.replace('<p>','')
    chapter_content = chapter_content.replace('</p>','')
    print(chapter_content)
    
```

```python
with open('%s.txt'%title,'w',encoding='utf-8') as file_project:
    for chapter_info in chapter_info_list[:32]:
        chapter_url,_,chapter_title = chapter_info
        #chapter_url = "https://read.dmnico.cn/book/401348.html%s"%chapter_url#用+号相连的话，会不停生成新的字符串变量，占内存
        chapter_response = requests.get(chapter_url)
        chapter_response.encoding = 'utf-8'
        chapter_html = chapter_response.text
        #提取章节内容
        chapter_content = re.findall(r'<div class="read-content j_readContent">(.*?)</div>',chapter_html,re.S)[0]
        #清洗数据
        chapter_content = chapter_content.replace(' ','')
        chapter_content = chapter_content.replace('<p>','')
        chapter_content = chapter_content.replace('</p>','')
        print(chapter_url,chapter_title)
        file_project.write(chapter_title)
        file_project.write('\n')
        file_project.write(chapter_content)
        file_project.write('\n')
```
