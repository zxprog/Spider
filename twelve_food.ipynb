```python
import requests # 调用requests库
from bs4 import BeautifulSoup # 调用BeautifulSoup库
from gevent import monkey
#从gevent库里导入monkey模块。
monkey.patch_all()
#monkey.patch_all()能把程序变成协作式运行，就是可以帮助程序实现异步。
import gevent,time,requests
#导入gevent、time、requests
from gevent.queue import Queue
#从gevent库里导入queue模块
from Analysis import *
import csv

```

```python
start = time.time()
url_menu = 'http://www.boohee.com/food/view_menu'
url_list = ['http://www.boohee.com/food/group/%d'%i for i in range(1,11)]
url_list.append(url_menu)
#print(url_list)

#c创建空队列
work = Queue()
for url in url_list:
    work.put_nowait(url)
count = 0
def crawler():
    global count
    with open('HeatQuality_of_food_2.csv','a', newline = '',encoding = 'gb18030') as csvfile:
        writer=csv.writer(csvfile,dialect='excel')
        writer.writerow(['食物名称','热量'])
        while not work.empty():
            url2 = work.get_nowait()
            for i in range(1,11):
                url_new = url2+'?'+'page={}'.format(i)
                soup = analysis(url_new)
                items = soup.find_all('li',class_='item clearfix')
                for item in items:
                    food_name = item.find('h4').text
                    heat_quality = item.find('p').text
                    print(food_name,heat_quality,url_new)
                    writer.writerow([food_name,heat_quality])
                    count+=1

tasks_list = []
#定义协程操作
for i in range(11):
    task = gevent.spawn(crawler)
    tasks_list.append(task)
gevent.joinall(tasks_list)

end = time.time()
print(end-start)
print(count)
```
